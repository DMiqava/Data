```python
import pandas as pd

df = pd.read_csv('survey_results_public.csv')
```

```python
desired_rows_and_columns = df.loc[[2, 4, 7], ['Country', 'Student']]
print("Desired rows and columns:")
print(desired_rows_and_columns)
```

```python
df.set_index('YearsCode', inplace=True) 
print("\nIndexed DataFrame:")
print(df)
```

```python
filter_condition = (df['Country'] == 'India') & (df['Age1stCode'] == '10')
filtered_df = df[filter_condition]
print("\nFiltered DataFrame:")
print(filtered_df)
```

```python
sorted_df = df.sort_values(by=['parameter1', 'parameter2'], ascending=[True, False])
print("\nSorted DataFrame:")
print(sorted_df)
```

```python
print("\nStatistics:")
print("Mean of Respondent:", df['Respondent'].mean())
print("Standard Deviation of Respondent:", df['Respondent'].std())
print("Median of Respondent:", df['Respondent'].median())
print("Minimum of Respondent:", df['Respondent'].min())
print("Maximum of Respondent:", df['Respondent'].max())
```
